Este notebook documenta e executa testes sequenciais nos endpoints da nossa API. Ele serve tanto como validação técnica quanto como documentação interativa para explicar o funcionamento e as regras de negócio do sistema.

```python
import requests
```

## 1. Teste de Saúde da API (Health Check)

**Objetivo:** Garantir que o servidor Spring Boot está rodando e respondendo a requisições HTTP básicas.

**Comportamento Esperado:** Retornar Status HTTP `200 OK` e o texto `hello` no corpo da resposta.

```python
url_health = 'http://localhost:8080/api/health'
response = requests.get(url_health)

print(f'Status Code: {response.status_code}')
print(f'Response Body: {response.text}')

assert response.status_code == 200, 'Erro: Status code não é 200'
assert response.text == 'hello', 'Erro: Resposta diferente de "hello"'
print('✅ Teste de Health Check concluído com sucesso!')
```
