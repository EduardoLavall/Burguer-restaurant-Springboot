# Testes de API - Burguer Restaurant

Este notebook documenta e executa testes manuais dos endpoints atuais da API. Ele serve como apoio para validar as rotas prontas durante o desenvolvimento.

```python
import requests

BASE_URL = "http://localhost:8081/api"
```

## 1. Teste de saúde da API

Objetivo: garantir que o backend está no ar e respondendo na rota correta.

```python
url_saude = f"{BASE_URL}/saude"
response = requests.get(url_saude)

print(f"Status Code: {response.status_code}")
print(response.json())

assert response.status_code == 200, "Erro: status code diferente de 200"
assert response.json()["status"] == "UP", "Erro: status de saude diferente de UP"
assert response.json()["servico"] == "burguer-restaurant", "Erro: nome do servico inesperado"
print("Teste de saude concluido com sucesso!")
```

## 2. Listar produtos iniciais

Objetivo: confirmar que a massa inicial de produtos em memória foi carregada.

```python
url_produtos = f"{BASE_URL}/produtos"
response = requests.get(url_produtos)
produtos = response.json()

print(f"Status Code: {response.status_code}")
print(f"Quantidade de produtos carregados: {len(produtos)}")
print(produtos)

assert response.status_code == 200, "Erro: falha ao listar produtos"
assert len(produtos) >= 4, "Erro: massa inicial de produtos nao foi carregada"
print("Listagem inicial de produtos validada com sucesso!")
```

## 3. Cadastrar novo produto

Objetivo: validar o endpoint de cadastro com os campos atuais do contrato.

```python
novo_produto = {
    "nome": "Brownie da Casa",
    "descricao": "Brownie de chocolate com casquinha crocante.",
    "preco": 12.50,
    "categoria": "doce",
    "disponibilidade": True,
    "imagem": None
}

response = requests.post(url_produtos, json=novo_produto)
produto_criado = response.json()
produto_id = produto_criado["id"]

print(f"Status Code: {response.status_code}")
print(produto_criado)

assert response.status_code == 201, "Erro: produto nao foi criado"
assert produto_criado["nome"] == novo_produto["nome"], "Erro: nome do produto incorreto"
assert produto_criado["categoria"] == novo_produto["categoria"], "Erro: categoria incorreta"
print("Cadastro de produto validado com sucesso!")
```

## 4. Alterar preco do produto

Objetivo: validar a atualizacao de preco usando a rota de patch.

```python
payload_preco = {
    "preco": 14.90
}

response = requests.patch(f"{url_produtos}/{produto_id}/preco", json=payload_preco)
produto_atualizado = response.json()

print(f"Status Code: {response.status_code}")
print(produto_atualizado)

assert response.status_code == 200, "Erro: preco nao foi atualizado"
assert produto_atualizado["preco"] == payload_preco["preco"], "Erro: preco retornado diferente do esperado"
print("Atualizacao de preco validada com sucesso!")
```

## 5. Remover produto

Objetivo: validar a rota de remocao e confirmar que o produto fica indisponivel.

```python
response_delete = requests.delete(f"{url_produtos}/{produto_id}")
response_lista = requests.get(url_produtos)
produtos_apos_remocao = response_lista.json()
produto_removido = next(produto for produto in produtos_apos_remocao if produto["id"] == produto_id)

print(f"Status delete: {response_delete.status_code}")
print(produto_removido)

assert response_delete.status_code == 204, "Erro: remocao do produto falhou"
assert produto_removido["disponibilidade"] is False, "Erro: produto deveria estar indisponivel"
print("Remocao logica de produto validada com sucesso!")
```
